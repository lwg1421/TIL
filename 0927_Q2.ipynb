# 2-1 그룹별 집계/요약하기
## Boston 데이터셋의 TAX 칼럼이 TAX 칼럼의 중위값보다 큰 데이터를 대상으로, CHAS 칼럼과 RAD 칼럼 순으로 그룹을 지은 후 각 그룹의 데이터 개수를 구하시오. 단, CHAS, RAD 칼럼별 데이터 개수는 COUNT라는 칼럼으로 출력합니다.

```python
import pandas as pd
```

```python
data = pd.read_csv("D:/boston.csv")
```

```python
data.head()
```

```python
#  TAX 칼럼이 TAX 칼럼의 중위값보다 큰 데이터 추출
a = data.loc[data["TAX"]>data["TAX"].median()]
a
```

```python
# CHAS, RAD 칼럼으로 그룹화하여 개수 새기
b = a.groupby(by = ["CHAS","RAD"])["TAX"].count()
```

```python
# 현재 b는 시리즈 타입이므로 데이터프레임으로 형변환
c = pd.DataFrame(b)
```

```python
# 컬럼이름변경
c.columns = ["COUNT"]
```

```python
c
```

# 2-5 표준 변환하기
## Boston 데이터셋의 DIS 칼럼을 표준화 척도로 변환한 후, 0.4보다 크면서 0.6보다 작은 값들에 대한 평균을 구하시오. 단, 소수점 셋째 자리에서 반올림하여 소수점 둘째 자리까지 출력하시오.

```python
data = pd.read_csv("D:/boston.csv")
```

```python
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
std_data = scaler.fit_transform(data)
std_data = pd.DataFrame(std_data,columns = data.columns)
std_data.head()
```

```python
answer = std_data.loc[(0.4 < std_data["DIS"]) & (std_data["DIS"]<0.6)]["DIS"].mean()
answer
```

```python
round(answer,2)
```

# 2-6 유니크한 값 구하기
## Boston 데이터셋의 전체 칼럼에 대해서 중복을 제거한 유니크한 값을 구한 후 칼럼별로 유니크한 값의 개수를 기준으로 평균값을 구하시오.

```python
data = pd.read_csv("D:/boston.csv")
```

```python
data.head()
```

```python
len(data["CRIM"].unique())
col_list = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT', 'MEDV']
```

```python
total = 0
for i in col_list:
    n = int(len(data[i].unique()))
    total = total + n
total
```

```python
int(total/len(col_list))
```
