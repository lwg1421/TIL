```python
import requests
import re
import warnings
from bs4 import BeautifulSoup as bs
import pandas as pd

# 불필요한 경고 출력을 방지합니다.
warnings.filterwarnings('ignore')
```

```python
def gotowok(url) :
    html = requests.get(url, verify = False).text
    soup = bs(html,"lxml")
    menu_html = soup.select("div.menu_list > div.item > div.menu_info > p:nth-child(3)")

    menu_text = []
    for i in menu_html:
        text = i.get_text()
        menu_text.append(text)
    menu_merge = ",".join(menu_text)
    menu_merge_replace = menu_merge.replace(" ","").replace("\n","").replace("*메인메뉴는시기에따라변경될수있습니다.","")

    # 메뉴명
    menu_name = menu_merge_replace.split(",")

    # 가게이름 리스트 생성
    restaurant_name = []
    restaurant_category = []

    for num in range(len(menu_name)):
        restaurant_name.append("고투웍")
        restaurant_category.append("중식")
    
    gotowok_df = pd.DataFrame({"restaurant_name":restaurant_name,"restaurant_category":restaurant_category,"menu_name":menu_name})
    gotowok_df.to_csv("C:/crawring/gotowok_df.csv", index = False, encoding='utf-8-sig')
    
    return(gotowok_df)
```

```python
def paikboy(url) :
    html = requests.get(url, verify = False).text
    soup = bs(html,"lxml")
    menu_html = soup.select("div.item > div.menu_info > div > p:nth-child(1)")
    menu_info_html = soup.select("div.item > div.menu_info > p:nth-child(3)")

    menu_text = []
    for i in menu_html:
        text = i.get_text()
        menu_text.append(text)
    menu_merge = ",".join(menu_text)
    menu_merge_replace = menu_merge.replace(" ","").replace("\n","").replace("*메인메뉴는시기에따라변경될수있습니다.","")

    # 메뉴설명
    menu_info = []
    for i in menu_info_html:
        text = i.get_text()
        menu_info.append(text.replace("\n",""))

    # 메뉴명
    menu_name = menu_merge_replace.split(",")

    # 가게이름 리스트 생성
    restaurant_name = []
    restaurant_category = []

    for num in range(len(menu_name)):
        restaurant_name.append("빽보이")
        restaurant_category.append("양식")
    
    paikboy_df = pd.DataFrame({"restaurant_name":restaurant_name,"restaurant_category":restaurant_category,"menu_name":menu_name,"menu_info":menu_info})
    paikboy_df.to_csv("C:/crawring/paikboy_df.csv", index = False, encoding='utf-8-sig')

    return(paikboy_df)
```
