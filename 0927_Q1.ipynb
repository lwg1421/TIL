# 1-2 결측치 확인하기
### Boston 데이터셋의 RM 칼럼에 대한 결측치를 평균값으로 대치한 후에 산출된 표준편차 값과 결측치를 삭제한 후에 산출된 표준편차 값의 차이를 구하시오. 단 최종적으로 출력하는 차이는 양수이다.

```python
data = pd.read_csv("D:/boston.csv")
```

```python
data.head()
```

```python
# 결측값을 평균으로 대치
a = data["RM"].fillna(data["RM"].mean())
```

```python
# 결측값 삭제
b = data["RM"].dropna()
```

```python
# 두 표준편차의 차이를 양수로 출력
abs(a.std()-b.std())
```

# 1-3 이상값 확인하기
### Boston 데이터셋의 ZN 칼럼을 대상으로 ZN값의 평균값에서 표준편차의 1.5배보다 크거나 작은 ZN 값의 합계를 구하시오

```python
boston = pd.read_csv("D:/boston.csv")
```

```python
boston.head()
```

```python
plus = boston["ZN"].mean() + 1.5*boston["ZN"].std()
minus = boston["ZN"].mean() - 1.5*boston["ZN"].std()
print(plus)
print(minus)
```

```python
boston[boston["ZN"]> plus]["ZN"].sum() + boston[boston["ZN"] < minus]["ZN"].sum()
```

# 1-4 사분위수 구하기
### Boston 데이터셋에서 CHAS 칼럼과 RAD 칼럼을 제외한 칼럼에 한해서 칼럼별 IQR값을 구하시오. 단 출력 구조는 2열이고 1열은 보스턴 데이터 세트의 칼럼 이름이 표시되어야 한다.
#### IQR : 3사분위수-1사분위수

```python
dataset = pd.read_csv("D:/boston.csv")
```

```python
dataset.head()
```

```python
dataset_1 = dataset.drop(columns = ["CHAS","RAD"])
```

```python
dataset_1.head()
```

```python
iqr = dataset_1.describe().loc[["25%","75%"]].T
iqr
```

```python
iqr["IQR"] = iqr["75%"]-iqr["25%"]
iqr = iqr.drop(columns = ["25%","75%"])
```

```python
iqr
```
