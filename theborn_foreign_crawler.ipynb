```python
url = "https://www.gseek.kr/member/rl/courseInfo/onCourseCsInfo.do?menuId=UB00&menuStep=2&pMenuId=OTOP&courseSeq=4720&courseCsSeq=1&courseCateCode=B200&eduTypeCode=&stuSeq="
html = requests.get(url,verify = False).text
soup = bs(html,"lxml")
```

```python
# 목차 내용 담을 리스트 list_contents 생성
list_contents = soup.select("#areaContainer > div.tab_content_area > div > table > tbody > tr > td.title.noline > span")

# 조회수 담을 리스트 list_watch 생성
list_watch = soup.select("#areaContainer > div.tab_content_area > div > table > tbody > tr > td.dp_all.view")
```

```python
# list_contents에서 텍스트만 추출하여 list_text 리스트에 담음
# 목차명 list_text
list_text = []
for contents in range(len(list_contents)):
    text = list_contents[contents].get_text()
    list_text.append(text)

# 조회수 watch_text
watch_list = []
for watch in range(len(list_watch)):
    text = list_watch[watch].get_text()
    watch_list.append(text)
    
```

```python
playlist_df = pd.DataFrame({"title":list_text, "count":watch_list})
playlist_df.set_index(np.arange(1,len(watch_list)+1,1))
playlist_df
```
