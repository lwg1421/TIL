# 행/열 합계
### - df.sum() 함수 사용
### - 행과 열의 합계를 구할때는 sum(axis=0/1) - axis는 0이 기본
### - 각 열의 합계를 구할때는 sum(axis=0)
### - 각 행의 합계를 구할때는 sum(axis=1)

```python
import numpy as np
import pandas as pd
```

```python
# 데이터프레임 생성
np.random.seed(1)
df=pd.DataFrame(np.random.randint(10,size=(4,8)))
df
```

```python
# 행별 합계
df["sum"] = df.sum(1)
df
```

```python
# 열별 합계
df.loc["sum"] = df.sum(axis = 0)
df
```

```python
# 총 합계 값 0으로 변경
df.loc["sum",["sum"]] = 0
```

```python
# 최종 데이터프레임
df
```
